```python
import numpy as np

from sktime.datasets import load_airline
from sktime.forecasting.naive import NaiveForecaster
from sktime.utils.plotting import plot_series
```

# data

```python
y = load_airline()
y
```

# forecast

```python
fcst = NaiveForecaster(strategy="last", sp=12)
fcst
```

# model fit

```python
# forecast horizon: predict next 36 months
fh = np.arange(1, 37)
# model fit
fcst.fit(y, fh = fh)
```

# model predict

```python
y_pred = fcst.predict()
y_pred
```

# result

```python
fig, ax = plot_series(y, y_pred, labels=["train", "pred"])
```
